### Pyplot API vs Object-Orient API
* Pyplot API -> Quickly
* OO API -> Advanced

### `Pyplot API`

```python
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x), color = "blue", linestyle = "dashed", label = "sin(x)")
plt.plot(x, np.cos(x), color = "red", label = "cos(x)")

plt.xlabel("X")
plt.ylabel("F(x)")

plt.title("Sin(x) & Cos(x)")

plt.axis([0, 10, -1, 1])
plt.axis("equal")

plt.legend();
```

### `OO API`

```python
#1 Prepare Data
x = [1,2,3,4]
y = [11,22,33,44]
#2 Setup plot
fg, ax = plt.subplots(figsize=(5,5))
#3 Plot the data
ax.plot(x,y)

ax.set(title = "A simple plot",
      xlabel = "x",
      ylabel = "f(x)")
```

## Most common types of Matplotlib plots
* line
* scatter
* bar
* subplots()

`line`

```python
x = np.linspace(0, 10, 100)
```

```python
## OO API
fig, ax = plt.subplots()
ax.plot(x, x**2)
```

`scatter`

```python
# Pyplot API
plt.scatter(x, np.exp(x));
```

```python
# OO API

rng = np.random.RandomState(0)

x = rng.randn(100)
y = rng.randn(100)

sizes = 200*rng.rand(100)
colors = rng.rand(100)
fig, ax = plt.subplots()
img1 = ax.scatter(x, y, s=sizes, c=colors, cmap="viridis", alpha=0.7)
fig.colorbar(img1);
```

`bar`
* Vertical
* Horizontal

```python
sort_drink_prices = {"Coke" : 10,
                     "Pepsi" : 12,
                     "Sprite" : 8}
fig, ax = plt.subplots()
ax.bar(sort_drink_prices.keys(), sort_drink_prices.values())
ax.set(title="Sort Drink Prices", ylabel="Price($)");
```

```python
fig, ax = plt.subplots()
ax.barh(list(sort_drink_prices.keys()), list(sort_drink_prices.values()));
ax.set(title="Sort Drink Prices", xlabel="Price($)");
```

`histogram`

```python
np.random.seed(42)
students_height = np.random.normal(170, 10, 250)



plt.hist(students_height, bins=30);
```

### Subplots

```python
# Option 1:
x = np.arange(-5, 5)
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2, ncols = 2, figsize = (10, 5))
ax1.plot(x, x**2)
ax1.plot(x, 5*x)

ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(sort_drink_prices.keys(), sort_drink_prices.values())
ax4.hist(students_height, bins=30);
```

```python
# Option 2:

fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (10, 5))
ax[0,0].plot(x, x**2)
ax[0,0].plot(x, 5*x)

ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(sort_drink_prices.keys(), sort_drink_prices.values())
ax[1,1].hist(students_height, bins=30);
```
