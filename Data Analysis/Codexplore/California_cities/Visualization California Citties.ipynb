```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("california_cities.csv")
data.head()
```

```python
plt.style.use("seaborn-darkgrid")
```

```python
latd = data["latd"]
longd = data["longd"]
sizes = data["area_total_km2"]
population = data["population_total"]

data["area_total_km2"].sort_values(ascending = False)

```

```python
plt.scatter(longd, latd, c = np.log10(population), cmap="viridis", s=sizes, alpha = 0.6)

plt.title("California's Citites")
plt.xlabel("Longtitude")
plt.ylabel("Latitude")

plt.colorbar(label="log$_{10}$(population)")
plt.clim(3,7)

area_range = [50,100,300,500]
for area in area_range:
    plt.scatter([],[], s=area, c='k', label=str(area) +"km$^2$", alpha=0.4)
    
plt.legend(labelspacing=1, title='City Area')
plt.axis("equal");
```
