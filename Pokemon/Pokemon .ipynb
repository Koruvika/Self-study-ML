```python
data.loc[(data["Type 1"] == "Bug") | (data["Type 2"] == "Bug")]
```

```python
for data_type in data_types:
    data.loc[(data["Type 1"] == data_type) | (data["Type 2"] == data_type)]
```

```python
total = np.array(data.loc[(data["Type 1"] == "Bug") | (data["Type 2"] == "Bug")].head(20)["Total"])
hp = np.array(data.loc[(data["Type 1"] == "Bug") | (data["Type 2"] == "Bug")].head(20)["HP"])
atk = np.array(data.loc[(data["Type 1"] == "Bug") | (data["Type 2"] == "Bug")].head(20)["Attack"])
dfs = np.array(data.loc[(data["Type 1"] == "Bug") | (data["Type 2"] == "Bug")].head(20)["Defense"])
name = np.array(data.loc[(data["Type 1"] == "Bug") | (data["Type 2"] == "Bug")].head(20)["Name"])
```

```python
plt.bar(name, hp)
plt.bar(name, atk, bottom=hp)
plt.bar(name, dfs, bottom=hp+atk)
plt.xticks(rotation=90);
```

```python
plt.barh(name, hp, label="HP", color="lightgreen")
plt.barh(name, atk, left=hp, label="ATK", color="red")
plt.barh(name, dfs, left=hp+atk, label="DEF")
plt.legend();
# plt.xticks(rotation=90);
```
